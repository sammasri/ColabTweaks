```python
import ipywidgets as widgets
import threading, time, IPython
from google.colab import output
from IPython.display import display 

bar1 = widgets.IntProgress(min= 0, max= 100, step= 5, layout= {'width':'100%'})
btn0 = widgets.Button(description= "Start", layout= {'width': '80px'})
btn1 = widgets.Button(description= "Kill", layout= {'width': '80px'})
btn_ui = widgets.Button(description="uiclicker", layout={'display':'none'})
threadtoggle = widgets.HBox([btn0, btn1, btn_ui])
label1 = widgets.Label(value= "")
container = widgets.VBox([bar1,threadtoggle, label1])

def updateui(val):
    bar1.notify_change({'name': 'value', 'type':'update'})

btn_ui.on_click(updateui)
output.eval_js("""function findclicker() {
 els = Array.prototype.slice.call(document.getElementsByTagName('button')).filter(el => el.textContent.trim() === 'uiclicker'.trim());
 if(els.length > 0){els[0].click();}}setInterval(findclicker, 300);""") 

def kill(obj):
    state.killed= True
    state.alive= False
    state.currentstate = 0
    btn0.description= "Start"
    state.thread.join()
    state.thread = None

btn1.on_click(kill)

class appstate:
    currentstate = 0
    bar = None
    label= None
    btn = None
    togglecounter = 0
    killed= False
    alive= False
    thread= None

state = appstate()
state.bar = bar1
state.label = label1
state.btn = btn0

def threadtoggle(obj):
    if not state.thread or not state.alive:
        state.thread = threading.Thread(target= t1test, args= (state,))
        state.thread.start()
        state.alive = True
        state.killed = False
        print('thread has gotten alive')

    state.togglecounter += 1
    if state.currentstate == 0:
        state.currentstate = 1
        state.label.value = f"App is on x{state.togglecounter}"
        state.btn.description = "Stop"
    else:
        state.currentstate = 0
        state.label.value = f"App is off x{state.togglecounter}"
        state.btn.description = "Start"

btn0.on_click(threadtoggle)
display(container)

def t1test(state0):
    while True:
        time.sleep(0.4)
        if state0.bar.value >= 100:
            state0.bar.value = 0
        if state0.currentstate == 1:
            state0.bar.value += 1
        if state0.togglecounter == 10 or state0.killed:
            print('process has been killed')
            break
    
```

```python
import ipywidgets as widgets
import threading, time, IPython, sys
from IPython.display import display
env_colab = 'google.colab' in sys.modules
if env_colab:
    from google.colab import output

b1 = widgets.IntProgress(min= 0, max= 100, step= 5, layout= {'width':'100%'})
b2 = widgets.IntProgress(min= 0, max= 100, step= 7, layout= {'width':'100%'})

btn = widgets.Button(description= "Start", layout= {'width': '80px'})
btn1 = widgets.Button(description= "Kill", layout= {'width': '80px'})
btn_ui = widgets.Button(description="uiclicker", layout={'display':'none'})
h1 = widgets.HBox([btn, btn1, btn_ui])
l1 = widgets.Label(value= "")
c1 = widgets.VBox([b1, b2, h1, l1])

def updateui(val):
    l1.notify_change({'name': 'value', 'type':'update'})
btn_ui.on_click(updateui)

if env_colab:
    output.eval_js("""function findclicker() {
    els = Array.prototype.slice.call(document.getElementsByTagName('button')).filter(el => el.textContent.trim() === 'uiclicker'.trim());
    if(els.length > 0){els[0].click();}}setInterval(findclicker, 300);""") 

def kill(obj):
    state.killed= True
    state.alive= False
    state.currentstate = 0
    btn.description= "Start"
    state.thread.join()
    state.thread = None

btn1.on_click(kill)

class appstate:
    currentstate = 0
    bar = None
    label= None
    btn = None
    togglecounter = 0
    killed= False
    alive= False
    thread= None

state = appstate()
state.b1 = b1
state.b2 = b2
state.label = l1
state.btn = btn

def h1(obj):
    if not state.thread or not state.alive:
        state.thread = threading.Thread(target= t1test, args= (state,))
        state.thread.start()
        state.alive = True
        state.killed = False
        print('thread got alive')

    state.togglecounter += 1
    if state.currentstate == 0:
        state.currentstate = 1
        state.label.value = f"App is on x{state.togglecounter}"
        state.btn.description = "Stop"
    else:
        state.currentstate = 0
        state.label.value = f"App is off x{state.togglecounter}"
        state.btn.description = "Start"

btn.on_click(h1)
display(c1)

def t1test(state0):
    while True:
        time.sleep(0.2)
        if state0.b2.value >= 100:
            state0.b2.value = 0
            state0.b1.value = 0
        if state0.currentstate == 1:
            state0.b2.value += 1
            state0.b1.value += 2
        if state0.togglecounter == 10 or state0.killed:
            print('process has been killed')
            break
    
```
